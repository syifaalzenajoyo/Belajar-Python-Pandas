## Membuat Data Frame

Data frame sample yang akan kita gunakan adalah data frame yang terdiri dari dua buah kolom, yaitu kolom `pemain` yang berisikan nama pemain yang berhasil mencetak goal dan kolom `goal` yang berisikan jumlah goal yang dicetak oleh pemain dalam satu sesi pertandingan. 

```python
d = {'pemain':['Budi', 'Joni', 'Iwan', 'Budi', 'Budi', 'Iwan', 'Asep', 'Joni'], 
     'goal':[2, 1, 3, 1, 1, 2, 2, 3]}
df = pd.DataFrame(d)
df
```

## Menampilkan Nilai Kumulatif (Cummulative)

- Pemanggilan `cumsum()` berfungsi untuk mendata jumlah goal secara kumulatif.

```python
df['goal'].cumsum().to_frame()
```

- Membuat kolom baru `'jumlah_goal_kumulatif'` yang akan menampung nilai `cumsum()` dari kolom `goal`.

```python
df['jumlah_goal_kumulatif'] = df['goal'].cumsum()
df
```

- Membuat kolom baru `'jumlah_goal_kumulatif_tiap_pemain'` yang akan menampung nilai `cumsum()` dari kolom `goal` yang dikelmpokkan berdasarkan kolom `pemain`.

```python
df['jumlah_goal_kumulatif_tiap_pemain'] = df.groupby('pemain')['goal'].cumsum()
df
```

- Membuat kolom baru `'cummax'` yang akan menampung nilai `cummax()` dari kolom `goal`. Pemanggilan `cummax()` akan mendata nilai tertinggi/maximum yang diperoleh sampai titik tertentu yaitu index. 

```python
df['cummax'] = df['goal'].cummax()
df
```

- Membuat kolom baru `'cummin'` yang akan menampung nilai `cummin()` dari kolom `goal`. Pemanggilan `cummin()` akan mendata nilai terkecil/minimum yang diperoleh sampai titik tertentu yaitu index. 

```python
df['cummin'] = df['goal'].cummin()
df
```

- Membuat kolom baru `'cumprod'` yang akan menampung nilai `cumprod()` dari kolom `goal`. Pemanggilan `cumprod()` (cummulative product) akan menghasilkan nilai hasil total perkalian kumulatif. 
    - Pada index ke-0, cumprod = 2
    - Pada index ke-1, cumprod = 2x1 = 2
    - Pada index ke-2, cumprod = 2x1x3 = 6
    - Pada index ke-3, cumprod = 2x1x3X1 = 6
    - Pada index ke-4, cumprod = 2x1x3X1x1 = 6
    - Pada index ke-5, cumprod = 2x1x3X1x1x2 = 12
    - Pada index ke-6, cumprod = 2x1x3X1x1x2x2 = 24
    - Pada index ke-7, cumprod = 2x1x3X1x1x2x2x3 = 72

```python
df['cumprod'] = df['goal'].cumprod()
df
```
