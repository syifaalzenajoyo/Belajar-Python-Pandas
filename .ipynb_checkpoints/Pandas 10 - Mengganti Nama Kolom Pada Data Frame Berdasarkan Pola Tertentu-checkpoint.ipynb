# Membuat Data Frame
1. Membuat data set *titanicfull.csv* menjadi format data frame dengan menggunakan method *pd.read_csv*. 
2. Mengatur nama kolom-kolom menjadi *'Pclass', 'Survival status', 'full Name', 'Sex ', ' Age','Sib SP', 'Parch', 'Ticket', 'Fare', 'Cabin',* dan *'Embarked'*.
3. Membuat backup data frame dengan melakukan deep copy. Untuk itu, kita memanggil method *df.copy(deep=True)*.
4. Menampilkan 5 baris dari hasil data frame yang telah dibuat menggunakan method *df.head()*.

```python
df = pd.read_csv('C:/Users/ASUS/Desktop/Belajar-Python-Pandas/titanicfull.csv')
df.columns = ['Pclass', 'Survival status', 'full Name', 'Sex ', ' Age',
             'Sib SP', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']
df_backup = df.copy(deep=True)

df.head()
```

#Melakukan modifikasi dan perbaikan terhadap kolom

- Menggunakan lowercase untuk nama kolom dan mengganti spasi dengan _

Kita akan mengubah nama kolom dari data frame sebelumnya, misalnya yaitu 'Pclass', 'Survival status', dan 'full Name' menjadi 'pclass', 'survival_status', dan 'full_name'.

Untuk melakukan hal tersebut, kita dapat memanggil *str.replace(' ', '_')* untuk mengganti spasi menjadi underscore (_) .

Untuk menerapkan standarisasi nama kolom dengan hanya menggunakan lowercase, kita dapat memanggil method *str.lower()*.

```python
df.columns = df.columns.str.replace(' ', '_').str.lower()
df.head()
```

#Memangkas kelebihan spasi pada nama kolom

Untuk melakukan standarisasi penggunaan lowercase untuk nama kolom, penggantian spasi menjadi '_', dan membuang spasi berlebih pada nama kolom seperti pada kolom 'sex ' dan ' age' , kita dapat melakukan hal seperti berikut ini.
- Membuat backup data frame dengan melakukan deep copy. Untuk itu, kita memanggil method *df.copy(deep=True)*.
- Mengonversi *df.columns* menjadi lowercase dengan menggunakan method *str.lower()*
- Membuang setiap kelebihan spasi baik di sisi kiri, maupun di sisi kanan, dengan menggunakan *str.strip()*.
- Mengganti spasi menjadi underscore (_) dengan memanggil *str.replace(' ', '_')*.

```python
df = df_backup.copy(deep=True)

df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')
df.head()
```
