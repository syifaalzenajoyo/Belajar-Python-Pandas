## Melakukan Pengaturan Tampilan

Untuk melakukan pengaturan tampilan, kita dapat memanggil `set_option` yang disertai oleh beberapa parameter.
- Untuk menentukan jumlah maksimal baris yang akan ditampilkan, kita dapat menyertakan parameter `'display.max_rows'`.
    > `pd.set_option('display.max_rows', 5)`. Menampilkan baris dengan jumlah maksimal 5 baris. 2 baris pertama paling atas, 2 baris terakhir paling bawah, dan 1 baris ... di tengah yang merepresentasikan baris lainnya.
- Untuk menentukan jumlah maksimal kolom yang akan ditampilkan, kita dapat menyertakan parameter `'display.max_columns'`.
    > `pd.set_option('display.max_columns', 6)`. Menampilkan kolom dengan jumlah maksimal 6 kolom. 3 kolom pertama di sebelah kiri, 3 kolom terakhir di sebalah kanan, dan 1 kolom ... di tengah yang merepresentasikan kolom lainnya.
- Untuk menentukan lebar maksimal dari kolom yang akan ditampilkan, kita dapat menyertakan parameter `'display.max_colwidth'`.
    > `pd.set_option('display.max_colwidth', 5)`. Menampilkan kolom dengan maksimal 20 karakter.

```python
pd.set_option('display.max_rows', 5)
pd.set_option('display.max_columns', 6)
pd.set_option('display.max_colwidth', 20)

df
```

## Mereset Display ke Kondisii Default

```python
pd.reset_option('^display.', silent=True)

df
```
