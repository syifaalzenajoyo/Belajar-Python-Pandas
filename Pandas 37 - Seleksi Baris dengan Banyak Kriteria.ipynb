## Seleksi Baris dengan Banyak Kriteria

Melakukan seleksi baris dengan data yang memenuhi 4 kriteria, yaitu :
    1. 'sex' = 'female'
    2. 'age' >= 60
    3. 'embarked' = 'S'
    4. 'survived' == 1

### Cara 1
- Setiap kriteria harus diapit oleh tanda kurung `()`.
- Menggunakan tanda `&` untuk menggabungkan semua kriteria.

```python
df[(df['sex']=='female') & (df['age']>=60) & (df['embarked']=='S') & (df['survived']==1)]
```

### Cara 2
- Setiap kriteria harus diapit oleh tanda kurung `()`.
- Menggunakan tanda `&` untuk menggabungkan semua kriteria.
- Setiap kriteria ditempatkan pada baris yang berbeda untuk memudahkan kita dalam membaca codenya.

```python
df[
    (df['sex']=='female') & 
    (df['age']>=60) & 
    (df['embarked']=='S') & 
    (df['survived']==1)
]
```

### Cara 3
- Membuat sebanyak n variabel yang akan menampung sebanyak n kriteria (setiap kriteria dimasukkan ke variabel terpisah)
- Memasukkan variabel pada seleksi baris dengan memanggil `df[kr1 & kr2 & kr3 & kr4]`

```python
kr1 = df['sex']=='female'
kr2 = df['age']>=60
kr3 = df['embarked']=='S'
kr4 = df['survived']==1

df[kr1 & kr2 & kr3 & kr4]
```
