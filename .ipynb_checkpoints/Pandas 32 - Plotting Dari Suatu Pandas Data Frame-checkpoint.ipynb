## Membuat sebuah Pandas Data Frame.

- Data frame yang akan dibuat memiliki 40 baris, 5 kolom, serta untuk setiap kolomnya diberi label A, B, C, D, dan E.

- Untuk mebuat labelnya, kita dapat menggunakan tuple yang dikenakan terhadap String.

- ex : String 'ABCDE' yang dikenakan/dikonversi ke tuple akan menghasilkan tuple 'A', 'B', 'C', 'D', dan 'E'.

> `pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)`

- `pd.DataFrame()` digunakan untuk membuat sebuah objek data frame

- `np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)` artinya data frame akan diisi oleh bilangan integer dengan jangkauan 1-20 (1,2,3,...,19) yang dipilih secara random. Kemudian, untuk size yang digunakan ditentukan oleh n_rows dan n_cols.

```python
n_rows = 40
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), 
                  columns=cols)
df.head()
```

## Line Plot

`df.head()` digunakan untuk menampilkan 5 baris pertama.

```python
df.head()
```

Untuk melakukan visualisasi data frame sebagai line plot, kita dapat memanggil `plot()` yang diikuti oleh parameter `kind='line'`.

#### Note:
Dihasilkan line plot yang merepresentasikan semua kolom.

```python
df.plot(kind='line')
```

Untuk menghasilkan line plot yang merepresentasikan kolom A dan B,
kita dapat seleksi kolomnya terlebih dahulu dengan `df[['A', 'B']]`. Lalu, kita dapat memanggil `plot()` yang diikuti oleh parameter `kind='line'`.

```python
df[['A', 'B']].plot(kind='line')
```

## Bar Plot

`df.head()` digunakan untuk menampilkan 5 baris pertama.

```python
df.head()
```

Untuk melakukan visualisasi data frame sebagai bar plot, kita dapat memanggil `plot()` yang diikuti oleh parameter `kind='bar'`.

```python
df.plot(kind='bar')
```

Untuk menghasilkan bar plot yang merepresentasikan kolom A dan B,
kita dapat seleksi kolomnya terlebih dahulu dengan `df[['A', 'B']]`. Lalu, kita dapat memanggil `plot()` yang diikuti oleh parameter `kind='bar'`.

```python
df[['A', 'B']].plot(kind='bar')
```

Untuk menghasilkan bar plot yang merepresentasikan kolom A dan B serta hanya menampilkan 5 baris pertamanya, kita dapat seleksi kolomnya terlebih dahulu dengan `df[['A', 'B']]`, lalu diikuti oleh `head()`. Setelah itu, kita dapat memanggil `plot()` yang diikuti oleh parameter `kind='bar'`.

```python
df[['A', 'B']].head().plot(kind='bar')
```

Untuk menghasilkan bar plot secara *horizontal* yang merepresentasikan kolom A dan B serta hanya menampilkan 5 baris pertamanya, kita dapat seleksi kolomnya terlebih dahulu dengan `df[['A', 'B']]`, lalu diikuti oleh `head()`. Setelah itu, kita dapat memanggil `plot()` yang diikuti oleh parameter `kind='barh'`.

```python
df[['A', 'B']].head().plot(kind='barh')
```

## Area Plot

`df.head()` digunakan untuk menampilkan 5 baris pertama.

```python
df.head()
```

Untuk melakukan visualisasi data frame sebagai area plot, kita dapat memanggil `plot()` yang diikuti oleh parameter `kind='area'`.

#### Note:
Nilai pada sumbu y merepresentasikan nilai kumulatif seluruh kolom pada baris tertentu. Misalnya pada baris dengan index 1.
- Koordinat sumbu y kolom A akan dimulai dari 0-5 (+5)
- Koordinat sumbu y kolom B akan dimulai dari 5-22 (+17)
- Koordinat sumbu y kolom C akan dimulai dari 22-37 (+15)
- Koordinat sumbu y kolom D akan dimulai dari 37-49 (+12)
- Koordinat sumbu y kolom E akan dimulai dari 49-57 (+8)

```python
df.plot(kind='area')
```

Untuk menghasilkan area plot yang merepresentasikan kolom A dan B serta hanya menampilkan 5 baris pertamanya, kita dapat seleksi kolomnya terlebih dahulu dengan `df[['A', 'B']]`, lalu diikuti oleh `head()`. Setelah itu, kita dapat memanggil `plot()` yang diikuti oleh parameter `kind='area'`.

```python
df[['A', 'B']].head().plot(kind='area')
```

## Scatter Plot

`df.head()` digunakan untuk menampilkan 5 baris pertama.

```python
df.head()
```

Untuk melakukan visualisasi data frame sebagai bar plot, kita dapat memanggil `plot()` yang diikuti oleh 3 parameter 
- Parameter pertama merepresentasikan kolom pertama yang akan diplotting yaitu `A`
- Parameter kedua merepresentasikan kolom KEDUA yang akan diplotting yaitu `B`
- Parameter ketiga `kind='scatter'` untuk melakukan visualisasi data frame sebagai scatter plot.

```python
df.plot(x='A', y='B', kind='scatter')
```
