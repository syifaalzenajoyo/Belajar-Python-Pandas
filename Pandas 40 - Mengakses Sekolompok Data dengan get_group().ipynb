## Mengakses Sekelompok Data yang Sudah Terkelompok dengan `get_group()`

- Melakukan pengelompokkan data berdasarkan kolom `sex` dengan memanggil `df.groupby('sex')`
- grouped_df akan memiliki 2 kelompok data, yaitu kelompok dengan `sex=female` dan `sex=male`

```python
grouped_df = df.groupby('sex')
```

- Mengakses kelompok data `sex=female` dan menampilkan 10 data pertamanya.

```python
grouped_df.get_group('female').head(10)
```

- Melakukan pengelompokkan data berdasarkan kolom `sex` dengan memanggil `df.groupby('survived')`
- grouped_df akan memiliki 2 kelompok data, yaitu kelompok dengan `survived==0` dan `survived==1`

```python
grouped_df = df.groupby('survived')
```

- Mengakses kelompok data `survived==1` dan menampilkan 10 data pertamanya.

```python
grouped_df.get_group(1).head(10)
```
