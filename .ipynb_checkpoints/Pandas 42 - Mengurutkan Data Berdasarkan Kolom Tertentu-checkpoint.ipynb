## Mengurutkan Data Berdasarkan Kolom Tertentu

- mengurutkan data dengan memanggil `sort_values()`dengan acuan kolom `age`. Ascendingnya bernilai True sehingga data diurutkan dari nilai terkecil-terbesar.

```python
df.sort_values('age').head()
```

- mengurutkan data dengan memanggil `sort_values()`dengan acuan kolom `age`. Ascendingnya bernilai False sehingga data diurutkan dari nilai terbesar-terkecil.

```python
df.sort_values('age', ascending=False).head()
```

- mengurutkan data dengan memanggil `sort_values()`dengan acuan kolom pertama yang diurutkan yaitu `survived` dan kolom kedua yaitu `age`.

```python
df.sort_values(['survived', 'age']).head()
```
